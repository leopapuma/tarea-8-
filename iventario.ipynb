```python
#librerias que se usaran
import sympy as sp
import numpy as np

# Parámetros del problema
c1 = 3
c2 = 2.5
d  = 187.5      # demanda diaria
h  = 0.02       # costo de almacenaje
K  = 20         # costo de ordenar
q  = 1000       # punto de descuento


# Ejercicio 1: función Q óptima

def Qoptimo(c1, c2, d, h, K, q):

    # escribimos  Q_0 sin descuento
    Q0 = np.sqrt((2*K*d)/h)

    # Función T(Q)
    def T(Q):
        if Q <= q:
            return c1*d + K*d/Q + h*Q/2
        else:
            return c2*d + K*d/Q + h*Q/2

    # Determinamos  en qué región caera
    if Q0 <= q:
        return Q0, "Región I (sin descuento)"

    # Si Q0 > q → Región II
    Q = sp.symbols('Q', real=True, positive=True)

    eq = sp.Eq(Q**2 + (2/h)*(c2*d - T(Q0))*Q + 2*K*d/h, 0)
    sol = sp.solve(eq, Q)

    # Solución válida: Q > q
    sol_validas = [float(s) for s in sol if s > q]

    if len(sol_validas) == 0:
        raise ValueError("No hay solución válida mayor que q")

    Q_star = sol_validas[0]
    return Q_star, "Región II (con descuento)"

# probamos la funcion
Q_star, region = Qoptimo(c1, c2, d, h, K, q)
Q_star, region
#respuesta
#Estás en la Región I sin descuento
# La política óptima es pedir 612.37 galones
```

```python
#ejercicio 2
def politica_inventario(c1, c2, d, h, K, q, L):

    # Obtener Q óptimo
    Q_star, region = Qoptimo(c1, c2, d, h, K, q)

    # Tiempo del ciclo
    t1 = Q_star / d

    # Punto de reorden
    R = L * d

    # Retraso efectivo
    retraso = np.floor(L / t1)

    return {
        "Q_optimo": Q_star,
        "Región": region,
        "Punto_de_reorden": R,
        "Tiempo_t1": t1,
        "Retraso_efectivo": int(retraso)
    }

# probamos la funcion
politica = politica_inventario(c1, c2, d, h, K, q, L=2)
politica
#Q_optimo = 612.37 unidades
#Este es el lote económico de compra (EOQ).
#Región = “Región I (sin descuento)”
#Punto_de_reorden = 375 unidades
#Tiempo_t1 = 3.266 días Es el tiempo entre pedidos
#Retraso_efectivo = 0
```
